```python
date_format = '%Y-%m-%d'

START_DATE = datetime.datetime.strptime('2023-03-09', date_format)
END_DATE = datetime.datetime.strptime('2023-10-06', date_format)

def usage_ratio_before(row):
    edu_date = row['교육일자']
    diff_dates = edu_date - START_DATE
    return row['이전'] / diff_dates.days

def usage_ratio_after(row):
    edu_date = row['교육일자']
    diff_dates = END_DATE - edu_date
    return row['질문수'] / diff_dates.days

def compare_str_split(text):
    text_list = [text[i:i+2] for i in range(0,len(text),2)]
    return list(set(text_list))

df_webinar = a  # 데이터 파일
df_webinar['비교 결과'] = '이후'
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 엑셀 파일 경로
file_paths = [
    "webinar_1_v2.xlsx",
    "webinar_2_v2.xlsx",
    "webinar_3_v2.xlsx"
]

# 파일을 DataFrame으로 읽기
dfs = [pd.read_excel(path) for path in file_paths]

# DataFrame들 결합
combined_df = pd.concat(dfs)
combined_df['CreatedDatetime'] = combined_df['CreatedDatetime'].apply(lambda x : x.split(' ')[0])
# 'CreatedDatetime' 열을 datetime 타입으로 변환
combined_df['CreatedDatetime'] = pd.to_datetime(combined_df['CreatedDatetime'])

# 'CreatedDatetime'으로 그룹화하고 각 날짜에 대한 고유한 사용자 수 계산
user_count_over_time = combined_df.resample('D', on='CreatedDatetime')['UserCode'].nunique()
# 특정 세 날짜에 해당하는 세로선 추가
# 선 그래프 그리기
plt.figure(figsize=(12, 6))
user_count_over_time.plot(kind='line', marker='o', color='b')
plt.title('일자별 고유 사용자 수', fontsize=14)
plt.xlabel('날짜', fontsize=12)
plt.ylabel('사용자 수', fontsize=12)
special_dates = ['2023-08-24', '2023-09-05', '2023-09-20']


for date in special_dates:
    date_obj = datetime.datetime.strptime(date, '%Y-%m-%d')
    plt.axvline(x=date_obj, color='red', linestyle='--', label=f'교육 일자: {date}')
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.tight_layout()

# 그래프 표시
plt.show()
```
